```python
# Fill the missing values in 'age' with the median age.
titanic_data["age"].fillna(titanic_data["age"].median(),inplace=True)
```

```python
# Handle missing data by dropping the 'deck' column
titanic_data.drop(columns=["deck"],inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python

titanic_data["embarked"].tail()
```

```python
# Impute the 'embarked' and 'embark_town' columns with the most common value
most_common_embarked=titanic_data["embarked"].mode()[0]
print(most_common_embarked)
```

```python
titanic_data["embarked"].fillna(most_common_embarked,inplace=True)
```

```python
most_common_embark_town=titanic_data["embark_town"].mode()[0]
```

```python
most_common_embark_town
```

```python
titanic_data["embark_town"].fillna(most_common_embark_town,inplace=True)
```

```python

# Verify that missing data has been handled
titanic_data.isnull().sum()
```

```python
import numpy as np
```

```python
#Outlier Detection
# Compute Z-scores for 'age' and 'fare'
mean_age=np.mean(titanic_data['age'])
std_dev_age=np.std(titanic_data['age'])
z_score=(titanic_data['age']-mean_age)/std_dev_age
outlier_age=titanic_data['age'][np.abs(z_score)>2.5]
```

```python
outlier_age
```

```python
mean_fare=np.mean(titanic_data["fare"])
std_dev_fare=np.std(titanic_data["fare"])
z_score_fare=(titanic_data["fare"]-mean_fare)/std_dev_fare
outlier_fare=titanic_data["fare"][np.abs(z_score_fare)>2.5]
```

```python
outlier_fare
```

```python
titanic_data=titanic_data.dropna(subset=["age"])
```

```python
#Calculate the first and third quartile
Q1=titanic_data["age"].quantile(0.25)
Q3=titanic_data["age"].quantile(0.75)
#IQR
IQR_age=Q3-Q1
lower_bound=Q1-1.5*IQR_age
Upper_bound=Q3+1.5*IQR_age

#outlier
outlier_age=titanic_data["age"][(titanic_data["age"]<lower_bound)|(titanic_data["age"]>Upper_bound)]
print(outlier_age)
```

```python
#Data Transformati
titanic_data["sex"].head(15)
```

```python
titanic_data["embarked"].tail(15)
```

```python
import pandas as pd
```

```python
encoded_features=pd.get_dummies(titanic_data[["sex","embarked"]])
```

```python
encoded_features
```

```python
titanic_data=pd.concat([titanic_data,encoded_features],axis=1)
```

```python
titanic_data
```

```python
titanic_data["fare"].head(10)
```

```python
#Modify 'fare' feature to 'log_fare' using natural logarithm
titanic_data["fare"]=np.log10(titanic_data["fare"]+0.1)
```

```python
titanic_data["fare"]
```

```python
# Create a new binary encoded feature 'embarked_southampt
# Create a new binary encoded feature 'embarked_southampton'
embark_town_col = pd.DataFrame([1 if i == 'Southampton' else 1 for i in titanic_data['embark_town']], columns=["embarked_southampton"])
```

```python
embark_town_col
```

```python
titanic_data["embark_town"]=pd.DataFrame([1 if i == 'Southampton' else 0 for i in titanic_data['embark_town']], columns=["embarked_southampton"])
```

```python
titanic_data.head(10)
```

```python
#One-hot encode categorical variables
titanic_preprocessed = pd.get_dummies(titanic_data, columns=['sex', 'embarked', 'class', 'who', 'adult_male', 'embark_town', 'alive', 'alone'], drop_first=True)
```

```python
titanic_preprocessed
```

```python
#Handle any NaN values by filling them with the mean of the column
titanic_preprocessed = titanic_preprocessed.fillna(titanic_preprocessed.mean())
```

```python
train_data, test_data = train_test_split(titanic_preprocessed, test_size=0.3, random_state=42)
```

```python
test_data
```

```python
train_data
```

```python
# Separate the target variable ("survived") 
x_train = train_data.drop("survived", axis=1)
x_test=test_data.drop("survived",axis=1)
y_train = train_data["survived"]
y_test=test_data["survived"]
```

```python
x_test
```

```python
# Initialize a Logistic Regression model
logistic=LogisticRegression(max_iter=1000)
```

```python
# Training the Logistic Regression
logistic.fit(x_train,y_train)
```

```python
predictions = logistic.predict(x_test) 
```

```python
predictions
```

```python
print("Accuracy:", accuracy_score(y_test, predictions))
print("\n Classification Report:\n")
print(classification_report(y_test, predictions))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions)) 
```

```python
print("Accuracy Score:")
print(accuracy_score(y_test, predictions))
```

```python
#Random FOres
from sklearn.ensemble import RandomForestClassifier
```

```python
RandomForest=RandomForestClassifier(n_estimators=100,random_state=42)
RandomForest.fit(x_train,y_train)
```

```python
y_pred_rf=RandomForest.predict(x_test)
```

```python
y_pred_rf
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\n Classification Report:\n")
print(classification_report(y_test, y_pred_rf))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf)) 
```

```python
print("Accuracy Score:")
print(accuracy_score(y_test, y_pred_rf))
```
